```python
import json
import os
import matplotlib.pyplot as plt
from datetime import datetime
```

```python

# Initialize lists to store the dates and counts for top comments and answers
dates = []
top_counts = []
answer_counts = []

# Loop through all the files in the "comments_json" folder
for filename in os.listdir("comments_json_anonym/"):
  # Check if the file is a json file
  if not filename.endswith(".json"):
    # Skip the file if it is not a json file
    continue

  # Extract the date from the file name
  date = datetime.strptime(filename.split("_")[0], "%Y-%m-%d")

  # Load the comments from the file
  with open(f"comments_json_anonym/{filename}", "r", encoding="utf-8") as f:
    comments = json.load(f)

  # Initialize counters for the number of top comments and answers
  top_count = 0
  answer_count = 0

  # Loop through all the comments and add up the total number of top comments and answers
  for comment in comments:
    top_count += 1
    if "answers" in comment:
      answer_count += len(comment["answers"])

  # Add the date and comment counts to the lists
  dates.append(date)
  top_counts.append(top_count)
  answer_counts.append(answer_count)

plt.figure(figsize=(20,6))
# fig, (ax1,ax2) = plt.subplots(nrows=2,ncols=1,figsize=(20,12),sharex=True)

# Create a bar plot with separate bars for top comments and answers
plt.bar(dates, answer_counts, bottom=top_counts, color="plum", label="Answers", width=0.8)
plt.bar(dates, top_counts, color="purple", label="Top comments", width=0.8)

# Rotate the x-axis labels to make them easier to read
plt.xticks(rotation=45)

# Add labels to the x-axis and y-axis
plt.xlabel("Date")
plt.ylabel("Number of comments")

# Add a legend to the plot
plt.legend()

# Increase the size of the plot

# Show the plot
plt.show()
```

```python
# find the special post dates in september
comments = [top+answer for top, answer in zip(top_counts,answer_counts)]
first = -1
last = -1
for i in range(len(dates)):
    if dates[i].month == 9 and first == -1:
        first = i
    if dates[i].month != 9 and first != -1 and last == -1:
        last = i

first_i = comments.index(max(comments[first:last]),first)
second_i = comments.index(max(comments[first_i+1:last]),first_i)
dates[first_i].strftime(format='%Y-%m-%d'), dates[second_i].strftime(format='%Y-%m-%d'), first_i, second_i
```

```python
import json
import os
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from datetime import datetime

filenames = os.listdir("comments_json_anonym")
filenames.sort()

post_lines = []

# Initialize the figure and the axis
fig, ax = plt.subplots(figsize=(20, 6))

# Function to update the plot for the next json file
def update(i):
  x = {datetime.strptime(filename.split("_")[0], "%Y-%m-%d"):0 for filename in filenames}

  filename = filenames[i]
  if os.path.splitext(filename)[1].endswith("json"):
    with open(f"comments_json_anonym/{filename}", "r",encoding="utf-8") as f:
      comments = json.load(f)
  else:
    return

  post_date = datetime.strptime(filename.split("_")[0], "%Y-%m-%d")
  
  for comment in comments:
    comment_date = datetime.fromtimestamp(comment["created_at"]).date()
    if comment_date in x:
      x[comment_date] += 1
    else:
      x[comment_date] = 1
    if "answers" in comment:
      for answer in comment['answers']:
        answer_date = datetime.fromtimestamp(answer["created_at"]).date()
        if answer_date in x:
          x[answer_date] += 1
        else:
          x[answer_date] = 1

  ax.clear()

  ax.bar(x.keys(), x.values(),color="plum",width=0.8)

  ax.axvline(x=post_date, color="purple")
  ax.set_ylim(0,250)

# Create the animation using the `FuncAnimation` function
ani = animation.FuncAnimation(fig, update, frames=range(len(filenames)), repeat=True)

plt.show()


```

```python
import json
import os
import matplotlib.pyplot as plt
from scipy.stats import linregress, pearsonr

# Initialize lists to store the counts of comments and likes
comments = []
likes = []

id = {}
i = 0
# Loop through all the json files in the "ichbinsophiescholl" folder
for filename in os.listdir("ichbinsophiescholl"):
  # Check if the file is a json file
  if filename.endswith(".json"):
    # Load the json file
    with open(f"ichbinsophiescholl/{filename}", "r") as f:
      data = json.load(f)
    id[i] = filename
    i += 1
    # Extract the count of comments and likes from the json data
    comments.append(data["node"]["edge_media_to_comment"]["count"])
    likes.append(data["node"]["edge_media_preview_like"]["count"])

plt.figure(figsize=(20,6))
# Create a scatter plot of the counts of comments and likes
plt.scatter(comments, likes,color="plum",marker='x')

# Calculate the slope, intercept, and other statistics for a linear regression
slope, intercept, r_value, p_value, std_err = linregress(comments, likes)

# Use the slope and intercept to create a list of y-values for the regression line
y_values = [slope * x + intercept for x in comments]

# Add the regression line to the scatter plot
plt.plot(comments, y_values, color="purple")

# Add labels to the x- and y-axes
plt.xlabel("Count of comments")
plt.ylabel("Count of likes")

# Calculate the Pearson correlation coefficient and the p-value for the correlation
corr, p_value = pearsonr(comments, likes)

# Print the Pearson correlation coefficient
print(f"Pearson correlation coefficient: {format(corr, '.10f')}")

# Print the p-value for the correlation
print(f"p-value for the correlation: {format(p_value,'.10f')}")

# Show the plot
plt.show()
```
